```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
url_student = 'https://raw.githubusercontent.com/Instantutor/Instantutor-Research/main/Datasets/Data.csv'
student = pd.read_csv(url_student)
url_rating = 'https://raw.githubusercontent.com/Instantutor/Instantutor-Research/main/Datasets/Rating.csv'
rating = pd.read_csv(url_rating)
student.head(3)
```

```python
rating.describe()
```

```python
rating_agg = rating.groupby('Mentor').Rating.agg(['count','mean']).reset_index()
#Rename columns
rating_agg = rating_agg.join(student['Area'])

print(rating_agg.head(5))
```

```python
#changing format of data
result = pd.DataFrame({"Mentor_ID":rating_agg['Mentor'], 
                       "Mean":rating_agg['mean'], 
                       "Count":rating_agg['count']})
result1 = result.pivot_table(index = 'Mentor_ID', columns = 'Count', values = 'Mean').fillna(0)
result1
```

```python
#compute cosine similarity
allusers = result1.values
user25629 = allusers[0]
denominator1 = np.sqrt(sum([np.square(x) for x in user25629]))

cosine_similarity = [(25629, 1)]
i = 1
for user in allusers[1:]:
  numerator = [x*y for x,y in zip(user25629,user)]
  denominator2 = np.sqrt(sum([np.square(x) for x in user]))
  costheta = sum(numerator)/(denominator1 * denominator2)
  cosine_similarity.append((result1.index[i], costheta))
  i+=1

cosine_similarity.sort(key=lambda x: x[1], reverse = True)
#display top 5 similar users
similar5users = cosine_similarity[0:5]
similar5users
```

```python
#map results

top5usersdf = pd.DataFrame()
top5usersdf
for user in similar5users:
  top5usersdf = top5usersdf.append(result1.loc[user[0]])
top5usersdf['costheta'] = [user[1] for user in similar5users]
all_values = top5usersdf.values
top5usersdf
```

```python
#predict ratings

denominator = sum([x[1] for x in similar5users])
inx = 0
values = []
for x in top5usersdf.loc[25629]:
  totalsum = 0
  if x == 0.0:
    for v in range(1, 5):
      totalsum += all_values[v-1][inx]*all_values[v-1][4]
    top5usersdf.loc[25629][inx+1] = totalsum/denominator
top5usersdf
```
